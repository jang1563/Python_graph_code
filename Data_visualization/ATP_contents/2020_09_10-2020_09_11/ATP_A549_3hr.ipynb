```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")
```

```python
data1 = pd.read_excel('ATP_control.xlsx')
data1.head()
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(22,12))

colors = ['#636363', '#2171b5', '#6baed6', '#bdd7e7']

sns.set_palette(sns.color_palette(colors))

sns.barplot(x="Conc. (nM)", y="ATP contents", hue="Time (hr)", data=data1, capsize=.1, ci="sd")
sns.swarmplot(x="Conc. (nM)", y="ATP contents", data=data1, hue="Time (hr)", color="0.2", size=15, dodge=True)


plt.legend(bbox_to_anchor=(1.08, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='40')

plt.xticks(size=40)
plt.yticks(size=40)


plt.xlabel("Conc. (nM)", size=40)
plt.ylabel("ATP contents", size=40)

plt.savefig('ATP_A549_time_1497.pdf', dpi = 300)
```
