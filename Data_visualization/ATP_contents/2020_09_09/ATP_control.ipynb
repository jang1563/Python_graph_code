```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")
```

```python
data1 = pd.read_excel('ATP_control.xlsx')
data1.head()
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(22,12))

colors = ['#636363', '#eff3ff', '#bdd7e7', '#6baed6', '#2171b5']

sns.set_palette(sns.color_palette(colors))

sns.barplot(x="Conc. (nM)", y="ATP contents", data=data1, capsize=.1, ci="sd")
sns.swarmplot(x="Conc. (nM)", y="ATP contents", data=data1, color="0.2", size=15, dodge=True)


plt.legend(bbox_to_anchor=(1.15, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='40')

plt.xticks(size=40)
plt.yticks(size=40)

plt.yscale('log')


plt.xlabel("Conc. (nM)", size=40)
plt.ylabel("ATP contents", size=40)

plt.savefig('ATP_contents_bar.pdf', dpi = 300)
```

```python
f, ax = plt.subplots(figsize=(18,10))

colors = ['#636363', '#eff3ff', '#bdd7e7', '#6baed6', '#2171b5']


sns.set_palette(sns.color_palette(colors))

sns.pointplot(x="Conc. (nM)", y="ATP contents", data=data1, ci="sd", scale = 2, markers=[8, u's', u'X', 11, u'x', u'P', 9, 6, 7],  errwidth=5.0, capsize=0.25)

plt.xticks(size=40)
plt.yticks(size=40)

plt.xlabel("Conc. (nM)", size=40)
plt.ylabel("ATP contents", size=40)

plt.legend(bbox_to_anchor=(1.17, 1.0))
plt.setp(ax.get_legend().get_texts(), fontsize='40')

plt.yscale('log')

plt.savefig('ATP_control_line.pdf', dpi = 300)
```
