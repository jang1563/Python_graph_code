```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")
```

```python
data1 = pd.read_excel('A549_1497.xlsx')
data1.head()
```

```python
data2 = pd.read_excel('A549_2136.xlsx')
data2.head()
```

```python
data3 = pd.read_excel('A549_2305.xlsx')
data3.head()
```

```python
data4 = pd.read_excel('A549_2422.xlsx')
data4.head()
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(22,12))

colors = ['#d9d9d9', '#2171b5', '#6baed6', '#bdd7e7']

sns.set_palette(sns.color_palette(colors))

sns.barplot(x="Cell line.1", y="ADP/ATP", hue="Conc. (nM)", data=data1, capsize=.1, ci="sd")
sns.swarmplot(x="Cell line.1", y="ADP/ATP", data=data1, hue="Conc. (nM)", color="0.2", size=15, dodge=True)


plt.legend(bbox_to_anchor=(1.11, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='40')

plt.xticks(size=40)
plt.yticks(size=40)


plt.xlabel("1497", size=40)
plt.ylabel("ADP/ATP \n normalized by control", size=40)

plt.savefig('ADP-to-ATP_A549_1497_6hr.pdf', dpi = 300)
```
