```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('Puro_H23.xlsx')
```

```python
data.head()
```

```python
sns.barplot(x="Conc. (ug/ml)", y="Cell confluence", data=data, capsize=.3, ci="sd")
sns.swarmplot(x="Conc. (ug/ml)", y="Cell confluence", data=data, color="0", alpha=.35)

plt.show()
```

```python
sns.barplot(x="Conc. (ug/ml)", y="Cell confluence", data=data, capsize=.3, ci="sd")
sns.swarmplot(x="Conc. (ug/ml)", y="Cell confluence", data=data, color="0", alpha=.35)

plt.title('Puromycin with H23', fontsize=20)
plt.savefig('Puromycin with H23', dpi = 300)
```
