```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data1 = pd.read_excel('Phenformin_72hr_data.xlsx')
data1.head()
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(16,12))

sns.barplot(x="Cell line", y="Cell confluence", data=data1, hue="Conc. (mM)", capsize=.1, ci="sd", palette="YlGnBu_d")
sns.swarmplot(x="Cell line", y="Cell confluence", data=data1, color="0.2", hue="Conc. (mM)", dodge=True)


plt.legend(bbox_to_anchor=(1.12, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='15')

plt.xticks(size=20, rotation=45,ha="right",rotation_mode='anchor')
plt.yticks(size=20)

plt.xlabel("Cell line", size=20)
plt.ylabel("Cell confluence (%)", size=20)

sns.set(style="whitegrid")

plt.title('Phenformin growth inhibition', fontsize=20)
plt.savefig('Phenformin growth inhibition 72 hr.pdf', dpi = 300)
```
