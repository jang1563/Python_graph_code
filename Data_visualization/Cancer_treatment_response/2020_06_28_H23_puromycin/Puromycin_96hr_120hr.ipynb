```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('Puromycin_analysis_120hr.xlsx')
data.head()
```

```python
sns.barplot(x="Conc. (ug/ml)", y="Cell confluence", data=data, hue="Cell line", capsize=.3, ci="sd")
sns.swarmplot(x="Conc. (ug/ml)", y="Cell confluence", data=data, color="0.2", hue="Cell line", dodge=True)
plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))

plt.title('Puromycin selection 120 hr', fontsize=20)
plt.savefig('Puromycin_selection_120hr_bar.pdf', dpi = 300)
```

```python
sns.barplot(x="Conc. (ug/ml)", y="Cell confluence", data=data, hue="Cell line", capsize=.3, ci="sd")
sns.swarmplot(x="Conc. (ug/ml)", y="Cell confluence", data=data, color="0.2", hue="Cell line", dodge=True)
plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
plt.xlim(1.0, 1.01)

plt.savefig('labe.pdf', dpi = 300)
```
