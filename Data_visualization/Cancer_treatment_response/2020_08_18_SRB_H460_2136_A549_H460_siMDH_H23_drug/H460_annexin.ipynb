```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('H460_Annexin_SRB.xlsx')
data.head()
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(22,12))

sns.barplot(x="Conc. (uM)", y="Confluence (%)", data=data, hue="Cell line", capsize=.1, ci="sd", palette="YlGnBu_d")
sns.swarmplot(x="Conc. (uM)", y="Confluence (%)", data=data, color="0.2", hue="Cell line", size=15, dodge=True)


plt.legend(bbox_to_anchor=(1.15, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='40')

plt.xticks(size=40)
plt.yticks(size=40)

plt.xlabel("Conc. (uM)", size=40)
plt.ylabel("Confluence (%)", size=40)

sns.set(style="whitegrid")

plt.title('LMT2136 Annexin V', fontsize=40)
plt.savefig('H460_LMT2136 apoptosis_confluence.pdf', dpi = 300)
```
