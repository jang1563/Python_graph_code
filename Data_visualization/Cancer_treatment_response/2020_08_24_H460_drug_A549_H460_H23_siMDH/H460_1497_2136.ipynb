```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")
```

```python
data1 = pd.read_excel('H460_1497.xlsx')
data1.head()
```

```python
data2 = pd.read_excel('H460_2136.xlsx')
data2.head()
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(22,12))

sns.barplot(x="Conc. (uM)", y="Cell confluence", data=data1, hue="Cell line", capsize=.1, ci="sd", palette="YlGnBu_d")
sns.swarmplot(x="Conc. (uM)", y="Cell confluence", data=data1, color="0.2", hue="Cell line", size=15, dodge=True)


plt.legend(bbox_to_anchor=(1.15, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='40')

plt.xticks(size=40)
plt.yticks(size=40)

plt.xlabel("Conc. (uM)", size=40)
plt.ylabel("Confluence (%)", size=40)

sns.set(style="whitegrid")

plt.savefig('H460_1497.pdf', dpi = 300)
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(22,12))

sns.barplot(x="Conc. (uM)", y="Cell confluence", data=data2, hue="Cell line", capsize=.1, ci="sd", palette="YlGnBu_d")
sns.swarmplot(x="Conc. (uM)", y="Cell confluence", data=data2, color="0.2", hue="Cell line", size=15, dodge=True)


plt.legend(bbox_to_anchor=(1.15, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='40')

plt.xticks(size=40)
plt.yticks(size=40)

plt.xlabel("Conc. (uM)", size=40)
plt.ylabel("Confluence (%)", size=40)

sns.set(style="whitegrid")

plt.savefig('H460_2136.pdf', dpi = 300)
```
