```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")
```

```python
data1 = pd.read_excel('H23_siMDH_20nM.xlsx')
data1.head()
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(22,12))

sns.barplot(x="Group", y="Cell confluence", data=data1, hue="Cell line", capsize=.1, ci="sd", palette="YlGnBu_d")
sns.swarmplot(x="Group", y="Cell confluence", data=data1, color="0.2", hue="Cell line", size=15, dodge=True)


plt.legend(bbox_to_anchor=(1.15, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='40')

plt.xticks(size=40)
plt.yticks(size=40)

plt.xlabel("Treatment", size=40)
plt.ylabel("Cell confluence (%)", size=40)

sns.set(style="whitegrid")

plt.ylim(80, 108)

plt.title('siMDH growth inhibition', fontsize=40)
plt.savefig('H23_siMDH_direct_20nM growth inhibition_normalize_conc_72hr.pdf', dpi = 300)
```
