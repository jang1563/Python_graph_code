```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('Puromycin_analysis.xlsx')
data.head()
```

```python
sns.barplot(x="Conc. (ug/ml)", y="Cell confluence", data=data, hue="Cell line", capsize=.3, ci="sd")
sns.swarmplot(x="Conc. (ug/ml)", y="Cell confluence", data=data, color="0.2", hue="Cell line", dodge=True)


plt.title('Puromycin selection 120 hr', fontsize=20)
plt.savefig('Puromycin_selection 120 hr bar.pdf', dpi = 300)
```
