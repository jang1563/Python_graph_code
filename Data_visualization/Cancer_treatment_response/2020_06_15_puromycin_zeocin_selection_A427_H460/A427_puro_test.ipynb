```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('Puro_A427.xlsx')
```

```python
data.head()
```

```python
sns.lineplot(x="Conc. (ug/ml)", y="Cell confluence", err_style = "bars", data=data, ci="sd")

plt.show()
```

```python
sns.scatterplot(x="Conc. (ug/ml)", y="Cell confluence", data=data, ci="sd")
sns.lineplot(x="Conc. (ug/ml)", y="Cell confluence", err_style = "bars", data=data, ci="sd")
sns.set(font_scale=2)

plt.show()
```

```python
sns.barplot(x="Conc. (ug/ml)", y="Cell confluence", data=data, capsize=.3, ci="sd")
sns.swarmplot(x="Conc. (ug/ml)", y="Cell confluence", data=data, color="0", alpha=.35)

plt.title('Puromycin with A427', fontsize=20)
plt.savefig('Puromycin with A427.pdf', dpi = 300)
```
