```python
import numpy as np
import pandas as pd



import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

data = pd.read_excel('H2009_drug.xlsx')
```

```python
data.head()
```

```python
df = data.pivot('Conc. (uM)', 'Drug', 'A540')

df.head()
```

```python
sns.heatmap(df, annot=True, fmt='.2f')
plt.title('Heatmap of H2009 cell', fontsize=20)
plt.show()
```
