```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('Media_condition.xlsx')
data.head()
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(,6))

sns.barplot(x="Cell line", y="Cell confluence", data=data, hue="FBS (%)", capsize=.1, ci="sd", palette="YlGnBu_d")
sns.swarmplot(x="Cell line", y="Cell confluence", data=data, color="0.2", hue="FBS (%)", dodge=True)


plt.legend(bbox_to_anchor=(1.0, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='15')

plt.xticks(size=20)
plt.yticks(size=20)

plt.xlabel("Cell line", size=20)
plt.ylabel("Cell confluence (%)", size=20)

sns.set(style="whitegrid")

plt.title('Growth', fontsize=20)
plt.savefig('Growth.pdf', dpi = 300)
```
