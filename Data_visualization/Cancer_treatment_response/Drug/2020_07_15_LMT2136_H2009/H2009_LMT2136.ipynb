```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('H2009_1st_2nd.xlsx')
data.head()
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(12,6))

sns.barplot(x="Cell line", y="Cell confluence", data=data, hue="Conc. (uM)", capsize=.1, ci="sd", palette="YlGnBu_d")
sns.swarmplot(x="Cell line", y="Cell confluence", data=data, color="0.2", hue="Conc. (uM)", dodge=True)


plt.legend(bbox_to_anchor=(1.0, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='15')

plt.xticks(size=20)
plt.yticks(size=20)

plt.xlabel("Cell line", size=20)
plt.ylabel("Cell confluence (%)", size=20)

sns.set(style="whitegrid")

plt.title('Growth inhibition', fontsize=20)
plt.savefig('LMT2136 growth inhibition_H2009_1st_2nd.pdf', dpi = 300)
```

```python
f, ax = plt.subplots(figsize=(12,6))

sns.barplot(x="Conc. (uM)", y="Cell confluence", data=data, hue="Cell line", capsize=.3, ci="sd", palette="YlGnBu_d")
sns.swarmplot(x="Conc. (uM)", y="Cell confluence", data=data, color="0.2", hue="Cell line", dodge=True)


plt.legend(bbox_to_anchor=(1.0, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='15')

plt.xticks(size=20)
plt.yticks(size=20)

plt.xlabel("Conc. (uM)", size=20)
plt.ylabel("Cell confluence (%)", size=20)

plt.title('Growth inhibition', fontsize=20)
plt.savefig('LMT2136 growth inhibition_H2009_1st_2nd_2.pdf', dpi = 300)
```
