```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('H23_LMT2136_analysis.xlsx')
data.head()
```

```python
ax = sns.barplot(x="Conc. (uM)", y="Cell confluence", data=data, hue="Cell seeding number", capsize=.3, ci="sd")
sns.swarmplot(x="Conc. (uM)", y="Cell confluence", data=data, color="0.2", hue="Cell seeding number", dodge=True)

plt.title('H23 drug treatment 24 hr', fontsize=20)
plt.savefig('H23 drug treatment 24 hr_bar_re.pdf', dpi = 300)
```

```python
ax = sns.barplot(x="Conc. (uM)", y="Cell confluence", data=data, hue="Cell seeding number", capsize=.3, ci="sd")
g = sns.swarmplot(x="Conc. (uM)", y="Cell confluence", data=data, color="0.2", hue="Cell seeding number", dodge=True)

g.legend_.remove()

plt.title('H23 drug treatment 24 hr', fontsize=20)
plt.savefig('H23 drug treatment 24 hr_bar_re.pdf', dpi = 300)
```

```python
sns.barplot(x="Conc. (uM)", y="Cell confluence", data=data, hue="Cell seeding number", capsize=.3, ci="sd")

plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0), fontsize='x-large')

plt.title('H23 drug treatment 24 hr', fontsize=20)
plt.savefig('H23 drug treatment 24 hr_bar_label.pdf', dpi = 300)
```

```python
ax = sns.barplot(x="Conc. (uM)", y="Cell confluence", data=data, hue="Cell seeding number", capsize=.3, ci="sd")
g = sns.swarmplot(x="Conc. (uM)", y="Cell confluence", data=data, color="0.2", hue="Cell seeding number", dodge=True)

g.legend_.remove()

plt.title('H23 drug treatment 24 hr', fontsize=20)
plt.xlabel('Conc. (uM)', fontsize=10)

```
