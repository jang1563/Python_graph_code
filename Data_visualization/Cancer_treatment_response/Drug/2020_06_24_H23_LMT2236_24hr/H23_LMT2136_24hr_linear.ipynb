```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('H23_LMT2136_analysis.xlsx')
data.head()
```

```python
sns.lmplot(x="Conc. (uM)", y="Cell confluence", hue="Cell seeding number", data=data)

plt.xlim(-2, 32)
sns.set(font_scale=1.5)

plt.title('H23 drug treatment 24 hr', fontsize=20)
plt.savefig('H23 drug treatment 24 hr.pdf', dpi = 300)
```
