```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('2305.xlsx')
data.head()
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(22,12))

sns.barplot(x="Cell line", y="Cell confluence", data=data, hue="Conc. (uM)", capsize=.1, ci="sd", palette="YlGnBu_d")
sns.swarmplot(x="Cell line", y="Cell confluence", data=data, color="0.2", hue="Conc. (uM)", size=15, dodge=True)


plt.legend(bbox_to_anchor=(1.0, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='40')

plt.xticks(size=40)
plt.yticks(size=40)

plt.xlabel("Cell line", size=40)
plt.ylabel("Cell confluence (%)", size=40)

sns.set(style="whitegrid")

plt.title('2305 growth inhibition', fontsize=40)
plt.savefig('2305 growth inhibition_normalize_cell_line.pdf', dpi = 300)
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(22,12))

sns.barplot(x="Conc. (uM)", y="Cell confluence", data=data, hue="Cell line", capsize=.1, ci="sd", palette="YlGnBu_d")
sns.swarmplot(x="Conc. (uM)", y="Cell confluence", data=data, color="0.2", hue="Cell line", size=15, dodge=True)


plt.legend(bbox_to_anchor=(1.2, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='40')

plt.xticks(size=30)
plt.yticks(size=30)

plt.xlabel("Conc. (uM)", size=40)
plt.ylabel("Cell confluence (%)", size=40)

sns.set(style="whitegrid")

plt.title('2305 growth inhibition', fontsize=30)
plt.savefig('2305 growth inhibition_normalize_conc.pdf', dpi = 300)
```
