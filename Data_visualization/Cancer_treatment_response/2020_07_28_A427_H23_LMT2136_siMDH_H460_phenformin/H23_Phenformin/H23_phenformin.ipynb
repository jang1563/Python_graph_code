```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data1 = pd.read_excel('H23_24hr.xlsx')
data1.head()
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(16,12))

sns.barplot(x="Cell line", y="Cell confluence", data=data1, hue="Conc. (mM)", capsize=.1, ci="sd", palette="YlGnBu_d")
sns.swarmplot(x="Cell line", y="Cell confluence", data=data1, color="0.2", hue="Conc. (mM)", size=10, dodge=True)


plt.legend(bbox_to_anchor=(1.16, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='30')

plt.xticks(size=40)
plt.yticks(size=40)

plt.xlabel("Cell line", size=40)
plt.ylabel("Cell confluence (%)", size=40)

sns.set(style="whitegrid")

plt.title('Phenformin growth inhibition', fontsize=40)
plt.savefig('Phenformin growth inhibition 24 hr.pdf', dpi = 300)
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(16,12))

sns.barplot(x="Conc. (mM)", y="Cell confluence", data=data1, hue="Cell line", capsize=.1, ci="sd", palette="YlGnBu_d")
sns.swarmplot(x="Conc. (mM)", y="Cell confluence", data=data1, color="0.2", hue="Cell line", size=15, dodge=True)


plt.legend(bbox_to_anchor=(1.2, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='30')

plt.xticks(size=40)
plt.yticks(size=40)

plt.xlabel("Conc. (mM)", size=40)
plt.ylabel("Cell confluence (%)", size=40)

sns.set(style="whitegrid")

plt.title('Phenformin growth inhibition', fontsize=40)
plt.savefig('Phenformin growth inhibition 24 hr_conc.pdf', dpi = 300)
```
