```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")
```

```python
data = pd.read_excel('H460_1497.xlsx')
data.head()
```

```python
f, ax = plt.subplots(figsize=(18,10))

colors = ['#636363', '#3182bd', '#9ecae1']


sns.set_palette(sns.color_palette(colors))

sns.pointplot(x="Conc. (uM)", y="Cell confluence", data=data, hue="Cell line", ci="sd", scale = 2, markers=[u's', 'o'],  linestyles=["--", "-"], errwidth=5.0, capsize=0.25)

plt.xticks(size=40)
plt.yticks(size=40)

plt.xlabel("Conc. (uM)", size=40)
plt.ylabel("Cell confluence (%)", size=40)

plt.legend(bbox_to_anchor=(1.17, 1.0))
plt.setp(ax.get_legend().get_texts(), fontsize='40')


plt.savefig('H460_1497_Empty_LKB1_line.pdf', dpi = 300)
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(22,12))

sns.barplot(x="Conc. (uM)", y="Cell confluence", data=data, hue="Cell line", capsize=.1, ci="sd", palette="YlGnBu_d")
sns.swarmplot(x="Conc. (uM)", y="Cell confluence", data=data, color="0.2", hue="Cell line", size=15, dodge=True)


plt.legend(bbox_to_anchor=(1.15, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='40')

plt.xticks(size=40)
plt.yticks(size=40)

plt.xlabel("Conc. (uM)", size=40)
plt.ylabel("Confluence (%)", size=40)

sns.set(style="whitegrid")

plt.savefig('H460_1497_bar.pdf', dpi = 300)
```
