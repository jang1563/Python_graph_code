```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('LMT2136_analysis.xlsx')
data.head()
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(22,12))

sns.barplot(x="Cell line", y="Cell confluence", data=data, hue="Conc. (uM)", capsize=.1, ci="sd", palette="YlGnBu_d")
sns.swarmplot(x="Cell line", y="Cell confluence", data=data, color="0.2", hue="Conc. (uM)", dodge=True)


plt.legend(bbox_to_anchor=(1.13, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='20')

plt.xticks(size=20, rotation=45,ha="right",rotation_mode='anchor')
plt.yticks(size=20)

plt.xlabel("Cell line", size=20)
plt.ylabel("Cell confluence (%)", size=20)

sns.set(style="whitegrid")

plt.title('LMT2136 growth inhibition', fontsize=20)
plt.savefig('LMT2136 growth inhibition_normalize_cell_line.pdf', dpi = 300)
```
