```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('H23.xlsx')
data.head()
```

```python
sns.axes_style("whitegrid")
```

```python
f, ax = plt.subplots(figsize=(12,6))

sns.pointplot(x="Time", y="Cell confluence", data=data, hue="Seeding number", ci="sd", palette="YlGnBu_d", markers=[8, u's', u'X', 11, u'x', u'P', 9, 6, 7], errwidth=0.8, capsize=0.5)

# Definition of tick_val and tick_lab
tick_val = [0, 6, 12, 18, 24, 30, 36, 42, 48, 54, 60, 66, 72]
tick_lab = ['0','12','24', '36', '48', '60', '72', '84', '96', '108', '120', '132', '144']

plt.xticks(tick_val,tick_lab, size=20)
plt.yticks(size=20)

plt.xlabel("Time", size=20)
plt.ylabel("Cell confluence (%)", size=20)

plt.legend(bbox_to_anchor=(1.0, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='15')

plt.title('H23 growth curve_resize', fontsize=20)
plt.savefig('H23 growth curve_resize.pdf', dpi = 300)
```
