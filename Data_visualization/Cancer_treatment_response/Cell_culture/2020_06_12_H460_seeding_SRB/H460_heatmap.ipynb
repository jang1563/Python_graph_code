```python
import numpy as np
import pandas as pd



import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline


data = pd.read_excel('H460_analysis.xlsx')
```

```python
data.head()
```

```python
df = data.pivot('Cell number', 'Time', 'A540')

df.head()
```

```python
sns.heatmap(df, annot=True, fmt='.1f')
plt.title('Heatmap of H460 cell', fontsize=20)
plt.show()
```

```python
sns.heatmap(df, fmt='.1f')
plt.title('Heatmap of H460 cell', fontsize=20)
plt.savefig('H460 heatmap_time', dpi = 300)
```
