```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('A549-48hr.xlsx')
```

```python
data.head()
```

```python
sns.regplot(x="Cell number", y="Cell confluence", data=data)
plt.xlim(0, 50000)

sns.set(font_scale=1.5)

plt.title('A549 confluence 48 hr', fontsize=20)
plt.savefig('A549 confluence 48 hr_regression.pdf', dpi = 300)
```
