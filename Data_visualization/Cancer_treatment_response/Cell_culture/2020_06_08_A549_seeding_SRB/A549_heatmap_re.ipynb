```python
import numpy as np
import pandas as pd



import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline


data = pd.read_excel('A549_analysis.xlsx')
```

```python
data.head()
```

```python
df = data.pivot('Cell number', 'Time', 'A540')

df.head()
```

```python
sns.heatmap(df, annot=True, fmt='.1f')
plt.title('Heatmap of A549 cell', fontsize=20)
plt.show()
```

```python
sns.heatmap(df, annot=True, fmt='.2f')
plt.title('Heatmap of A549 cell', fontsize=20)
plt.savefig('A549 heatmap annotation_time_dpi', dpi = 300)
```

```python
sns.heatmap(df, fmt='.1f')
plt.title('Heatmap of A549 cell', fontsize=20)
plt.savefig('A549 heatmap_time', dpi = 300)
```
