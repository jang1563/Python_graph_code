```python
import numpy as np
import pandas as pd



import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline


data = pd.read_excel('H23_analysis.xlsx')
```

```python
data.head()
```

```python
df = data.pivot('Cell number', 'Time', 'Cell confluence')

df.head()
```

```python
sns.heatmap(df, annot=True, fmt='.1f')
plt.title('Heatmap of H23 cell', fontsize=20)
plt.show()
```
