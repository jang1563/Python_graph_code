```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('H23_analysis_reg.xlsx')
data.head()
```

```python
sns.lmplot(x="Cell number", y="Cell confluence", hue = "Time", data=data)

plt.xlim(0, 50000)

sns.set(font_scale=1.5)

plt.title('H23 confluence', fontsize=20)
plt.savefig('H23 confluence reg.pdf', dpi = 300)
```
