```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('24_48_72_triplicate.xlsx')
data.head()
```

```python
ax = sns.barplot(x="Conc. (uM)", y="RLU", data=data, hue="Time", capsize=.3, ci="sd")
sns.swarmplot(x="Conc. (uM)", y="RLU", data=data, color="0.2", hue="Time", dodge=True)

plt.legend(bbox_to_anchor=(1.0, 1.0))

plt.title('Glucose consumption', fontsize=20)
plt.savefig('H23_Glucose_consumption_24_48_72_triplicate.pdf', dpi = 300)
```

```python
ax = sns.barplot(x="Time", y="RLU", data=data, hue="Conc. (uM)", capsize=.1, ci="sd")
sns.swarmplot(x="Time", y="RLU", data=data, color="0.2", hue="Conc. (uM)", dodge=True)

plt.legend(bbox_to_anchor=(1.0, 1.0))

plt.title('Glucose consumption', fontsize=20)
plt.savefig('H23_Glucose_consumption_24_48_72_RLU_triplicate.pdf', dpi = 300)
```
