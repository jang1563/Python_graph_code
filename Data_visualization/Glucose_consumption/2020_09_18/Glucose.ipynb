```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")
```

```python
data1 = pd.read_excel('A549_drug.xlsx')
data1.head()
```

```python
data2 = pd.read_excel('H460_drug.xlsx')
data2.head()
```

```python
sns.axes_style("whitegrid")

f, ax = plt.subplots(figsize=(22,12))

colors = ['#d9d9d9', '#2171b5', '#6baed6', '#bdd7e7']

sns.set_palette(sns.color_palette(colors))

sns.barplot(x="Drug", y="Glucose consumed", hue="Conc. (uM)", data=data1, capsize=.1, ci="sd")
sns.swarmplot(x="Drug", y="Glucose consumed", data=data1, hue="Conc. (uM)", color="0.2", size=15, dodge=True)


plt.legend(bbox_to_anchor=(1.11, 1.0))

plt.setp(ax.get_legend().get_texts(), fontsize='40')

plt.xticks(size=40)
plt.yticks(size=40)


plt.xlabel("Drug", size=40)
plt.ylabel("Glucose consumed", size=40)

plt.savefig('Glucose_A549_drug_6hr.pdf', dpi = 300)
```
