```python
import seaborn as sns, matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

sns.set(style="whitegrid")


data = pd.read_excel('30uM_triplicate.xlsx')
data.head()
```

```python
ax = sns.barplot(x="Conc. (uM)", y="RLU", data=data, hue="Cell number", capsize=.3, ci="sd")
sns.swarmplot(x="Conc. (uM)", y="RLU", data=data, color="0.2", hue="Cell number", dodge=True)

plt.legend(bbox_to_anchor=(1.0, 1.0))

plt.title('Glucose consumption', fontsize=20)
plt.savefig('H23_Glucose_consumption_30uM_triplicate.pdf', dpi = 300)
```

```python
ax = sns.barplot(x="Cell number", y="RLU", data=data, hue="Conc. (uM)", capsize=.1, ci="sd")
sns.swarmplot(x="Cell number", y="RLU", data=data, color="0.2", hue="Conc. (uM)", dodge=True)

plt.legend(bbox_to_anchor=(1.0, 1.0))

plt.title('Glucose consumption', fontsize=20)
plt.savefig('H23_Glucose_consumption_30uM_RLU_triplicate_consumption.pdf', dpi = 300)
```
